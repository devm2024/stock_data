```python
tickers_to_find = ' '.join(all_tickers.Symbol)
```

```python
data = yf.download(tickers_to_find, start="2020-01-01", end="2020-04-30", interval= '1d', group_by='ticker')
```

```python
data.columns = [' '.join(col).strip() for col in data.columns.values]
```

```python
data.columns = data.columns.get_level_values(0)
```

```python
data.head()
```

```python
data_close = data.filter(regex=("\w Adj Close"))
```

```python
data.to_csv('stocks_prices.csv', index=False)
```
